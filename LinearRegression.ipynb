```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
class LinearRegression:
    
    def __init__(self):
        self.w = None
    
    def fit(self, X, y):
        X = np.c_[np.ones((X.shape[0], 1)), X]
        self.w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
        
    def predict(self, X):
        X = np.c_[np.ones((X.shape[0], 1)), X]
        return X.dot(self.w)
```

```python
raw_df = pd.read_csv("Real_estate.csv", index_col=0)
data = raw_df.values[::, :-3]
target = raw_df.values[::, -1:]
```

```python
raw_df.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, train_size=0.8, shuffle=True)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Mean deviation", np.sum(abs(y_pred-y_test))/len(y_pred), "\n")
print("Predict           Real       Deviation")
for i in range(len(y_pred)):
    print(y_pred[i], '   ', y_test[i], '   ', abs(y_pred[i]-y_test[i]))
```
